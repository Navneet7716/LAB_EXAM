```python
import pulp as p
```

```python
M=4
N=4
NAV = M+N
a =range(1,M+1)
a1 = range(M)
b = range(1,N+1)
b1 = range(N)
xindx = [(a[i],b[j]) for j in b1 for i in a1]
tindx = range(1,NAV+1)
model = p.LpProblem('Assignment Problem' , p.LpMinimize)
x = p.LpVariable.dicts("X",xindx,None)
y = p.LpVariable.dicts("Y",tindx,None)

model += 13.0*x[1,1] + 16.0*x[1,2] + 12.0 * x[1,3] + 11.0*x[1,4] \
+ 15.0*x[2,1] + 2.0*x[2,2] + 13.0*x[2,3] + 20.0 *x[2,4]\
+ 5.0*x[3,1] + 7.0*x[3,2] + 10.0*x[3,3] +6.0*x[3,4] , "Assignment Cost"

model += x[1,1]+x[2,1]+x[3,1]+x[4,1]-y[1]>=1
model += x[1,2]+x[2,2]+x[3,2]+x[4,2]-y[2]>=1
model += x[1,3]+x[2,3]+x[3,3]+x[4,3]-y[3]>=1
model += x[1,4]+x[2,4]+x[3,4]+x[4,4]-y[4]>=1




model += x[1,1]+x[1,2]+x[1,3]+x[1,4]-y[5]>=1
model += x[2,1]+x[2,2]+x[2,3]+x[2,4]-y[6]>=1
model += x[3,1]+x[3,2]+x[3,3]+x[3,4]-y[7]>=1
model += x[4,1]+x[4,2]+x[4,3]+x[4,4]-y[8]>=1


print(model)
model.solve()

print("Status: " , p.LpStatus[model.status])

for v in model.variables():
    print(v.name,"=",v.varValue)
print("Objective Function: ", value(model.objective))
```
